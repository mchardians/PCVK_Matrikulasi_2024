```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as img
```

**Konvolusi Tanpa Library**

```python
def convolution2d(image, kernel, stride, padding):
  image_width, image_height = image.shape
  kernel_width, kernel_height = kernel.shape

  width = (image_width - kernel_width) // stride + 1
  height = (image_height - kernel_height) // stride + 1

  output = np.zeros((width, height)).astype(np.float32)

  for i in range(0, width):
    for j in range(0, height):
      output[i, j] = np.sum(image[i * stride:i * stride + kernel_width, j * stride:j * stride + kernel_height] * kernel).astype(np.float32)

  return output
```

**Operasi Sharpen**

```python
img_gray = cv.imread('Images/female.jpg')
img_gray = cv.cvtColor(img_gray, cv.COLOR_BGR2GRAY)

kernel_sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

cv2_imshow(convolution2d(img_gray, kernel_sharpen, 1, 2))
```

**Operasi Emboss**

```python
kernel_sharpen = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])

cv2_imshow(convolution2d(img_gray, kernel_sharpen, 1, 2))
```

**Operasi LEft Sobel Edge Detection**

```python
kernel_sharpen = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])

cv2_imshow(convolution2d(img_gray, kernel_sharpen, 1, 2))
```

**Operasi Canny Edge Detection**

```python
kernel_sharpen = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

cv2_imshow(convolution2d(img_gray, kernel_sharpen, 1, 2))
```

**Operasi Prewitt Edge Detection**

```python
kernel_sharpen = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

cv2_imshow(convolution2d(img_gray, kernel_sharpen, 1, 2))
```

**Operasi 5 x 5 Gaussian Blur**

```python
kernel_size = 5
sigma = math.sqrt(kernel_size)
gaussian_kernel = cv.getGaussianKernel(kernel_size, sigma)
gauss_kernel = gaussian_kernel @ gaussian_kernel.transpose()

cv2_imshow(convolution2d(img_gray, gauss_kernel, 1, 2))
```

**Operasi 21 x 21 Gaussian Blur**

```python
kernel_size = 21
sigma = math.sqrt(kernel_size)
gaussian_kernel = cv.getGaussianKernel(kernel_size, sigma)
gauss_kernel = gaussian_kernel @ gaussian_kernel.transpose()

cv2_imshow(convolution2d(img_gray, gauss_kernel, 1, 2))
```
